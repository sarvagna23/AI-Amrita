```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[51,42,53,44,54]
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(x,y,color='green', alpha=1, linewidth=5,linestyle='-.')
```

**Format strings in plot function**

```python
plt.plot(x,y,'g+')
plt.plot(x,y,'rD--')
plt.plot(x,y,'g+', markersize=20)
plt.plot(x,y,alpha=0.5) #alpha is for transparency
```

**Axes labels, legends, grid**

```python
days=[1,2,3,4,5]
max=[51,52,53,54,55]
min=[45,46,47,48,49]
avg=[47,50,52,53,54]
```

```python
plt.xlabel('days')
plt.plot(days,max, label='max')

plt.plot(days,min, label='min')
plt.plot(days,avg, label='avg') # this will make print them all in one chart
plt.legend(loc="best", shadow="True", fontsize="small") # it is used to mention where the labels of three graphs to be placed
# there are othr prop like upper_right, lower_right..
plt.grid() # grid makes it more easie to understand the graph
```

**Bar Chart**

```python
comp=['GOOG','AMZN','MSFT','FB','TAK']
rev=[90,70,87,79,45]
profit=[40,20,30,50,23]
```

```python
import numpy as np
```

```python
ypos= np.arange(len(comp))
ypos
#it is an numpy array
# as it wont allow string values

```

```python
plt.xticks(ypos,comp) #this will replace numpy values of an array to company names
# we can xlabel and ylabel and title as we have seen earlier
bars=plt.bar(ypos,rev, label='revenue')
#now to plot two graphs
plt.legend()
plt.show()
```

**Hatching or patterns filling in bargraphs**

```python
plt.xticks(ypos,comp) #this will replace numpy values of an array to company names
# we can xlabel and ylabel and title as we have seen earlier
plt.bar(ypos,rev, label='revenue',hatch='xx')
plt.bar(ypos,profit, label='profit',hatch='*')
#now to plot two graphs
plt.legend()
plt.show()
```

```python
#now to make a seperate list
xpos= np.arange(len(comp))
xpos
```

```python
plt.xticks(ypos,comp) #this will replace numpy values of an array to company names
# we can xlabel and ylabel and title as we have seen earlier
plt.bar(xpos-0.4,rev,width=0.2, label='revenue') # we hav added width
plt.bar(xpos+0.4,profit, width=0.2, label='profit')
#now to plot two graphs
plt.legend()
```

```python
plt.yticks(ypos,comp) #this will replace numpy values of an array to company names
# we can xlabel and ylabel and title as we have seen earlier
plt.barh(xpos-0.4,rev, label='revenue') #for horizontal bar
plt.barh(xpos+0.4,profit, label='profit')
#now to plot two graphs
plt.legend()
```

**Histograms**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,85,120,150,146,140,98]
plt.hist(blood_sugar, rwidth=0.8) # without mentioning bins default there will be 10
```

```python
blood_sugar=[113,85,120,150,146,140,98]
plt.hist(blood_sugar, bins=3, rwidth=0.8)

plt.show()
```

```python
blood_sugar=[113,85,120,150,146,140,98]
plt.hist(blood_sugar, bins=[80,100,140,160], rwidth=0.8, color='g',histtype='step') #rwidth is nothing but relative width of a bar compared to bin size.
plt.text(2.5,4,'Sarvagna_Analysis')
```

**Bin and Bucket**
by default histogram prints 10 bins and these binas are nothing but ranges.


**Now plotting two types of histograms side by side**

```python
blood_sugar=[113,85,120,150,146,140,98]
blood_sugar2=[83,97,110,130,145,70,155]
plt.hist([blood_sugar,blood_sugar2], bins=3, rwidth=0.8, color=['green','pink'], # here just a minor chnage we pass the two dats in  list
         label=['1','2'],orientation='horizontal') 
# by label parameter first chart for first data and second chart for second data
#orientation is to plot histogram as mentioned here it is given as horizontal
plt.ylabel('Total count')
plt.xlabel('Sugar range')
plt.legend()
plt.show()
```

**Pie chart**

```python
exp_val=[1400,600,300,410,250]
exp_labels=["Home rent","Food","Phone Bill","Car","Other Utilities"]
plt.pie(exp_val,labels=exp_labels)
```

```python
exp_val=[1400,600,300,410,250]
exp_labels=["Home rent","Food","Phone Bill","Car","Other Utilities"]
plt.pie(exp_val,labels=exp_labels)

plt.axis=("equal") # to make qual length of x and y axis
```

```python
# Increase radius of piechart
plt.pie(exp_val,labels=exp_labels, radius=1.5)
plt.show()
```

*autopct argument to show percentage**

```python
plt.pie(exp_val,labels=exp_labels, autopct='%0.2f%%',shadow='True') # %f%% is a format string and here we mention what format we want percentage in
# here it is in two dec points
plt.show()
```

Explode array

```python
plt.pie(exp_val,labels=exp_labels, autopct='%0.2f%%',shadow='True', explode=[0,0.2,0,0,0])
plt.show()
```

start angle argument: it makes the angle where the angle is started to where we mentioned.

```python
plt.pie(exp_val,labels=exp_labels, autopct='%0.2f%%',shadow='True', explode=[0,0.2,0,0,0],startangle=90)
plt.show()
```
